```python
i = 0
for music in url :
    response = get(music)
    html_soup = BeautifulSoup(response.text, 'html.parser')
    lyrics_con = html_soup.find_all("div", {"style" : "text-align: center;"})  
    lyrics_con = lyrics_con[0]
  #  print(lyrics_con.b.text[10:])
#     print(lyrics_con.text.find('\n\n\n'))
  #  print(lyrics_con.text[lyrics_con.text.find('\n\n\n')+3:])
    name = lyrics_con.b.text[10:].replace("\n", "")
    song_names.append(name)
    lyrics = lyrics_con.text[lyrics_con.text.find('\n\n\n')+3:]
    if 'ศิลปิน' in lyrics :
        lyrics = lyrics[lyrics.find('\n\n\n')+3:]
    if 'เนื้อร้อง' in lyrics:
        lyrics = lyrics[:lyrics.find('เนื้อร้อง')]
    lyrics = lyrics.replace("\n", " ")
    lyrics_list.append(lyrics)
    if i < 10:
        types.append('pop')
    elif  i < 20:
        types.append('rock')
    elif  i < 23:
        types.append('dance')
    elif  i < 25:
        types.append('modern')
    elif  i < 27:
        types.append('aucustic')
    elif  i < 31:
        types.append('Country')
    elif i == 31:
        types.append('Electronic')
    elif i  < 33:
        types.append('folk') 
    elif i  == 34:
        types.append('fank') 
    elif i < 39:
        types.append('rap')
    else:
        types.append('rock')
    i = i+1
print( song_names)
    
```

```python
import pandas as pd 
music_dict = {'Song name': song_names , 'Lyrics': lyrics_list, 'Genre':types }  
df = pd.DataFrame(music_dict)
df
```
